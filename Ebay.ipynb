```python
def snakecase(series):
    snake_case = []
    for s in series:
        if s.islower() == False:
            new_word = ''
            for c in s:              
                if c.islower():
                    new_word = new_word + c
                else:
                    c = c.lower()
                    new_word = new_word + "_" + c
            snake_case.append(new_word)
            
        else:
            snake_case.append(s)
            
    return(snake_case)
            
```

# Created a function to make the columns go from camelcase to snake case to be able to better read the columns. I created the function because there might be times where there are more than 100 columns and changing them individually would take too long. Rewrite this sectkion

```python
new_column = snakecase(autos.columns)

autos.columns = new_column
autos.columns
```

# After looking at the data, we see that both the seller and offer_type column both have the same value for each role. Thus we do not need them and will drop them. Also to take into consideration is that both the price and odometer columns are not saved as numeric values but text so these will be converted. 

```python
autos = autos.drop(["num_photos","seller","offer_type"], axis = 1)

autos["price"] = (autos["price"]
                                 .str.replace("$", "")
                                 .str.replace(",","")
                                 .astype(int)
                                 )

autos["odometer"] = (autos["odometer"].str.replace("km","")
                                      .str.replace(",","")
                                      .astype(int))

autos.rename({"odometer":"odometer_km"}, axis = 1 ,inplace = True)
```

We are going to stick with cars that were made in after 1900 as the table has vechiles that were made before then which is unlikely. We are also not going to surpass 2016 since that was when the data was collected

```python
autos = autos[autos["registration_year"].between(1900,2016)]
```

```python
autos["brand"].value_counts(normalize= True).sort_values(ascending = False
                                                       )
brand_values = autos["brand"].value_counts(normalize= True).sort_values(ascending = False
                                                       )
selected_brand = brand_values[brand_values > .05].index
```

```python
selected_brand
```

```python
selected_brands_mean = {}
for brands in selected_brand:
    brand = autos[autos["brand"] == brands]
    brand_mean = brand["price"].mean()
    brand_mean = int(brand_mean)
    selected_brands_mean[brands] = brand_mean
print(selected_brands_mean)
```

# we'll round the values to whole numbers to make it more readable. We can see that audi is the most expensive car brand

# slide working on mileage 

```python
selected_brand_avg_mileage= {}
for brands in selected_brand:
    brand = autos[autos["brand"] == brands]
    avg_mileage = brand["odometer_km"].mean()
    avg_mileage = int(avg_mileage)
    selected_brand_avg_mileage[brands] = avg_mileage
print(selected_brand_avg_mileage)
```

```python
brand_mean_prices_series = pd.Series(selected_brands_mean)
print(brand_mean_prices_series)
df = pd.DataFrame(brand_mean_prices_series, columns=["avg_price"])
df
```
